```python
# Map class labels to 0 (suicide) and 1 (non-suicide)
dataset.loc[dataset['class'] == 'suicide', 'class'] = 0
dataset.loc[dataset['class'] == 'non-suicide', 'class'] = 1
dataset['class'] = dataset['class'].astype(int)
print("Mapped classes:", dataset['class'].unique())
```

```python
# Text preprocessing with stemming
port_stem = PorterStemmer()

def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]', ' ', content)
    stemmed_content = stemmed_content.lower().split()
    stemmed_content = [port_stem.stem(word) for word in stemmed_content if word not in stopwords.words('english')]
    return ' '.join(stemmed_content)

dataset['stemmend_content'] = dataset['text'].apply(stemming)
```

```python
# TF-IDF vectorization
vectorizer = TfidfVectorizer(
    max_features=10000,
    min_df=5,
    max_df=0.90,
    ngram_range=(1, 3),
    sublinear_tf=True,
)
X_tfidf = vectorizer.fit_transform(dataset['stemmend_content'])
#keyword_flags = dataset['keyword_flag'].values
```

```python
X_combined = hstack([X_tfidf, keyword_flags.reshape(-1, 1)])
Y_label = dataset['class'].astype(int)
```

```python
print("Feature matrix shape:", X_combined.shape)
print("Labels shape:", Y_label.shape)
```

```python

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(X_combined, Y_label, test_size=0.2, random_state=52)
print("x_train:", x_train.shape, "x_test:", x_test.shape)
print("y_train:", y_train.shape, "y_test:", y_test.shape)

# Ensure int type for training
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
# 🔍 Hyperparameter tuning for SVM
from sklearn.svm import SVC
param_grid_svm = {
    'C': [0.65],              # Regularization strength
    'kernel': ['linear'],    # Linear or non-linear kernel
    'gamma': ['scale']      # Gamma value (used with rbf)
}

svm = SVC()
grid_search_svm = GridSearchCV(estimator=svm, param_grid=param_grid_svm, cv=5, scoring='accuracy', verbose=1)
grid_search_svm.fit(x_train, y_train)
```

```python
# Best model from grid search
best_svm_model = grid_search_svm.best_estimator_
print("✅ Best Hyperparameters for SVM:", grid_search_svm.best_params_)
```

```python
# Training accuracy
train_preds_svm = best_svm_model.predict(x_train)
score_train_svm = accuracy_score(y_train, train_preds_svm)
print("📈 Training Accuracy (SVM):", score_train_svm)
```

```python
# Testing accuracy
test_preds_svm = best_svm_model.predict(x_test)
score_test_svm = accuracy_score(y_test, test_preds_svm)
print("🧪 Testing Accuracy (SVM):", score_test_svm)
```

```python
# Evaluation
print("\n📊 Classification Report:\n", classification_report(y_test, test_preds_svm))
print("🧮 Confusion Matrix:\n", confusion_matrix(y_test, test_preds_svm))
```

```python
from sklearn.model_selection import learning_curve
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt

# Initialize SVM model
svm = SVC(kernel='linear', gamma='scale')  # You can adjust kernel or other hyperparameters

# Optional: use a subset of data for faster execution
X_sample = x_train[:500]
y_sample = y_train[:500]

# Generate learning curve data
train_sizes, train_scores, test_scores = learning_curve(
    svm,
    X_sample,
    y_sample,
    cv=3,
    scoring='accuracy',
    train_sizes=np.linspace(0.1, 1.0, 5),
    n_jobs=-1
)

# Calculate mean scores
train_scores_mean = np.mean(train_scores, axis=1)
test_scores_mean = np.mean(test_scores, axis=1)

# Plot 1: Training Accuracy
plt.figure(figsize=(10, 5))
plt.plot(train_sizes, train_scores_mean, color='blue', marker='o', label='Training Accuracy')
plt.title('Training Accuracy vs Training Set Size (SVM)')
plt.xlabel('Training Set Size')
plt.ylabel('Training Accuracy')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot 2: Validation Accuracy
plt.figure(figsize=(10, 5))
plt.plot(train_sizes, test_scores_mean, color='green', marker='o', label='Validation Accuracy')
plt.title('Validation Accuracy vs Training Set Size (SVM)')
plt.xlabel('Training Set Size')
plt.ylabel('Validation Accuracy')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import learning_curve
import matplotlib.pyplot as plt
import numpy as np

# Initialize SVM with probability=True for log loss
svm = SVC(kernel='linear', gamma='scale', probability=True)

# Use a subset of training data
X_small = x_train[:1000]
y_small = y_train[:1000]

# Learning curve for Log Loss
train_sizes, train_scores, test_scores = learning_curve(
    svm,
    X_small,
    y_small,
    cv=3,
    scoring='neg_log_loss',
    train_sizes=np.linspace(0.1, 1.0, 5),
    n_jobs=-1
)

# Convert to positive log loss
train_loss = -np.mean(train_scores, axis=1)
test_loss = -np.mean(test_scores, axis=1)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_loss, label='Training Log Loss', marker='o', color='blue')
plt.plot(train_sizes, test_loss, label='Validation Log Loss', marker='o', color='green')
plt.title('Loss Curve (Log Loss) for SVM (Subset)')
plt.xlabel('Training Set Size')
plt.ylabel('Log Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
